# Análisis de Tweets de la Cuenta @Claudiashein

## Introducción

Este reporte analiza 100 tweets recientes de la cuenta de Twitter @Claudiashein, enfocándose en estadística descriptiva básica, distribuciones de palabras, y n-gramas. Para ello, se implementó un código en Python utilizando bibliotecas como nltk, wordcloud, y tweepy. Además, se aplicó un filtro de stopwords para mejorar la relevancia del análisis.

## Metodología

**Obtención de datos:** Los datos fueron recolectados usando la API v2 de Twitter.

**Procesamiento de texto:**
- Limpieza de texto para eliminar enlaces, menciones, hashtags y caracteres especiales.

- Tokenización de palabras.

- Eliminación de stopwords en español y palabras con menos de tres caracteres.

**Análisis:**
- Cálculo de frecuencias de palabras.

- Identificación de los n-gramas más frecuentes.

- Visualización mediante nubes de palabras.
