```python
import pandas as pd

# Ruta del archivo CSV
ruta_archivo = '/content/AirBNBReviews.csv'

# Leer el archivo CSV
df = pd.read_csv(ruta_archivo)

# Mostrar el contenido del DataFrame
df.head(10)
```

```python
lexicon = pd.read_table(
            'https://raw.githubusercontent.com/fnielsen/afinn/master/afinn/data/AFINN-en-165.txt',
            names = ['termino', 'sentimiento']
          )
lexicon.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.stem.snowball import SnowballStemmer

def tokenizar(txt):
    # Check if the input is a string before applying lower()
    if isinstance(txt, str):
        texto_limpio = re.sub(r'[^\w\s]', '', txt.lower()) # Eliminamos signos de puntuación y ponemos todo en minúsculas
    else:
        # Handle non-string values, e.g., by returning an empty list
        return []

    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(texto_limpio)
    texto_filtrado = [word for word in tokens if word not in stop_words]
    return texto_filtrado
```

```python
df['token'] = df['Review'].apply(lambda x : tokenizar(x))
df.head(10)
```

```python
df['Review number'] = range(0,df.shape[0])
df.head(10)
```

```python
df_tokens_rev = df.explode(column='token').drop(columns='Review')
df_tokens_rev.head(20)
```

```python
df_sentimiento = pd.merge(
                  left     = df_tokens_rev,
                  right    = lexicon,
                  left_on  = "token",
                  right_on = "termino",
                  how      = "inner"
                 )

df_sentimiento.head(20)
```

```python
import numpy as np

def perfil_sentimientos(df):
    print(df.reset_index().loc[0,'Review number'])
    print("=" * 12)
    print(f"Positivos: {round(100 * np.mean(df.sentimiento > 0), 2)}")
    print(f"Neutros  : {round(100 * np.mean(df.sentimiento == 0), 2)}")
    print(f"Negativos: {round(100 * np.mean(df.sentimiento < 0), 2)}")
    print(" ")
```

```python
aggregated_data = df_sentimiento.groupby(['Review number']).agg(
    Real_Sentiment=('Positive or Negative', 'max'),
    Predicted_Sentiment=('sentimiento', 'mean')
)

aggregated_data['Predicted_Sentiment'] = aggregated_data['Predicted_Sentiment'].apply(lambda x: 1 if x > 0 else 0)
aggregated_data.head(10)
```

```python
# Comparar las dos columnas
aggregated_data['Match'] = aggregated_data['Real_Sentiment'] == aggregated_data['Predicted_Sentiment']

# Calcular la proporción de coincidencias
proporcion_coincidencia = aggregated_data['Match'].mean()

print(f"Proporción de coincidencia: {proporcion_coincidencia:.2%}")
```
